```python
#########################################################
################### WORD GUESSER ########################
#########################################################
## Get letter Function for user input##

def getLetter():
 
  # Create a variable to hold the letter with a dummy value
  letter = ""
 
  # Until the user enters a string with ONE character...
  while(len(letter) != 1):
 
    # Prompt the user to enter a letter
    letter = input("Enter a Letter: ")
 
  # Return the letter
  return letter

## Word Display Function for guessed letters in the word ##

  def displaySecretWord(word, guesses):
 
  for letter in word:
 
    # If this letter (from the secret word) has been guessed, display the letter
    if(letter in guesses):
      print(letter, end=" ")
    # Otherwise, display an underscore ( _ )
    else:
      print("_", end=" ")

### A function that determines if the user has won the game ###
def hasUserWon(word, guesses):
 
  # Let's use an "Innocent Until Proven Guilty" Algorithm...
  # ...and create a variable that is set to "True"
  won = True
 
  # Go through each letter in the secret word
  for letter in word:
 
    # Check if the letter has been guessed
    if(letter not in guesses):
 
      # If it has NOT been guessed, set the variable we created to False, and stop the loop
      won = False
      break
 
  return won

# Create the secret word
secretWord = "BUBBLES"
 
# Setup a collection to store user guesses
myGuess = ""
myGuesses = []

# Setup a variable to hold the number of strikes the user receives
strikes = 0
maxStrikes = 5
i = 0

# If the user has NOT won OR lost, take a turn
while((not hasUserWon(secretWord,myGuesses)) and (strikes < maxStrikes)):
  
  # Get a SINGLE letter from the user
  myGuess = getLetter()

  # Store the guess
  myGuess = myGuess.capitalize()
  myGuesses.append(myGuess)
  
  # Determine if this letter is in the secret word
  if(myGuess in secretWord):

  # If it is, display the secret word with all of our current guesses
    displaySecretWord(secretWord, myGuesses)
    print("Good Guess!")

  # If it isn't, give the user a strike and tell them that the letter is an incorrect guess
  else: 
    i += 1
  print(f"You have {i} strike. If you get 5, you lose!")

# If the user won, display that message, and the entire secret word
if (hasUserWon(secretWord, myGuesses)):
  print("You have won!")

# If the user lost, display that message
else:
  print("You have lost!")
```
